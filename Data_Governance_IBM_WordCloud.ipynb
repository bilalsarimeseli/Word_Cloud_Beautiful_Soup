Let's start by importing the necessary libraries. We'll need `requests` to fetch the web page content and **BeautifulSoup** to parse the HTML.

```python
import requests
from bs4 import BeautifulSoup
```

We need to fetch the web page content using the requests library that I want to create a word cloud from.

I will replace `URL_TO_WEB_PAGE` with the actual URL of the web page.

```python
URL_TO_WEB_PAGE = "https://www.ibm.com/topics/data-governance"  # Replacing my own URL
response = requests.get(URL_TO_WEB_PAGE)
web_page_content = response.text
```

Let's parse the **HTML** content of the web page using **BeautifulSoup** which allows us to extract the text from the page.

```python
soup = BeautifulSoup(web_page_content, "html.parser")
#To extract the text from the parsed HTML, I will use the get_text() method on the BeautifulSoup object.
web_page_text = soup.get_text()
```

```python
#Install & Import Wordcloud Library:
!pip install wordcloud
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Generate Word Cloud from the extracted text.
wordcloud = WordCloud(width=600, height=400, background_color="white").generate(web_page_text)
plt.figure(figsize=(15, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()

```
